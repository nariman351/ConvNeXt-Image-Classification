```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
classes = ['Airplane', 'Car', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']
gen_actual, gen_pred = np.loadtxt(open("class_pred_gen.csv"), delimiter=",", skiprows=1, unpack=True)
test_actual, test_pred = np.loadtxt(open("class_pred_test.csv"), delimiter=",", skiprows=1, unpack=True)
```

```python
def make_matrix(labeled_actual, labeled_predicted, dataset):
    plt.clf()
    c_matrix = confusion_matrix(labeled_actual, labeled_predicted)
    c_df = pd.DataFrame(c_matrix, index=classes, columns=classes)
    plt.figure(figsize=(8,8))
    sns.heatmap(c_df, annot=True, fmt='g')
    plt.ylabel('Actual Class')
    plt.xlabel('Predicted Class')
    plt.savefig(f'remake_cm_{dataset}.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
make_matrix(gen_actual, gen_pred, 'gen')
make_matrix(test_actual, test_pred, 'test')
```
